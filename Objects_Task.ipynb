```python
from ultralytics import YOLO
from paddleocr import PaddleOCR
import cv2
import json

multi_model = YOLO('yolov8n.pt')
plate_model = YOLO('best.pt')
ocr = PaddleOCR(use_angle_cls=True, lang='en')
```

```python
import matplotlib.pyplot as plt

img_path = '/content/138-1-8.webp'
img = cv2.imread(img_path)

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

```python
results_multi = multi_model(img)

img_with_boxes = results_multi[0].plot()
plt.imshow(cv2.cvtColor(img_with_boxes, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
for result in results_multi:
    for box in result.boxes:
        cls_id = int(box.cls[0])
        conf = float(box.conf[0])
        label = multi_model.names[cls_id]
        print(f"{label}: {conf:.2f}")
```

```python
results_plates = plate_model(img)

img_with_plates = results_plates[0].plot()
plt.imshow(cv2.cvtColor(img_with_plates, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
for result in results_plates:
    for box in result.boxes:
        cls_id = int(box.cls[0])
        conf = float(box.conf[0])
        label = plate_model.names[cls_id]
        print(f"{label}: {conf:.2f}")
```

```python
results_plates = plate_model(img)

for result in results_plates:
    boxes = result.boxes.xyxy.cpu().numpy()
    for i, box in enumerate(boxes):
        x1, y1, x2, y2 = map(int, box)
        plate_img = img[y1:y2, x1:x2]
        ocr_result = ocr.predict(plate_img)

        text, confidence = "", 0.0
        if ocr_result and len(ocr_result) > 0:
            rec_texts = ocr_result[0].get('rec_texts', [])
            rec_scores = ocr_result[0].get('rec_scores', [])
            if rec_texts and rec_scores:
                text = rec_texts[0]
                confidence = rec_scores[0]

        print(f"Plate {i+1}: Text = {text}, Confidence = {confidence}")
```

```python
results_multi = multi_model(img)
scene_objects = []
for res in results_multi:
    for box, cls in zip(res.boxes.xyxy.cpu().numpy(), res.boxes.cls.cpu().numpy()):
        x1, y1, x2, y2 = map(int, box)
        label = multi_model.names[int(cls)]
        scene_objects.append({
            "label": label,
            "bbox": [x1, y1, x2, y2]
        })
```

```python
ocr_result = ocr.predict(plate_img)

if ocr_result and len(ocr_result) > 0:
    rec_texts = ocr_result[0].get('rec_texts', [])
    rec_scores = ocr_result[0].get('rec_scores', [])
    if rec_texts and rec_scores:
        text = rec_texts[0]
        confidence = rec_scores[0]
    else:
        text = ""
        confidence = 0.0
else:
    text = ""
    confidence = 0.0

print(f"Text: {text}, Confidence: {confidence}")
```

```python
results_plates = plate_model(img)
plates_info = []

for result in results_plates:
    boxes = result.boxes.xyxy.cpu().numpy()
    for i, box in enumerate(boxes):
        x1, y1, x2, y2 = map(int, box)
        plate_img = img[y1:y2, x1:x2]

        ocr_result = ocr.predict(plate_img)

        text, confidence = "", 0.0
        if ocr_result and len(ocr_result) > 0:
            rec_texts = ocr_result[0].get('rec_texts', [])
            rec_scores = ocr_result[0].get('rec_scores', [])
            if rec_texts and rec_scores:
                text = rec_texts[0]
                confidence = rec_scores[0]

        plate_data = {
            "bbox": [x1, y1, x2, y2],
            "text": text,
            "confidence": confidence
        }
        plates_info.append(plate_data)
```

```python
from collections import Counter
counts = Counter([obj['label'] for obj in scene_objects])
scene_description = ", ".join([f"{v} {k}(s)" for k,v in counts.items()])

output = {
    "scene_description": scene_description,
    "detected_objects": scene_objects,
    "license_plates": plates_info
}

print("Scene Description:", scene_description)
print(json.dumps(output, indent=2))
```

```python
from transformers import Blip2Processor, Blip2ForConditionalGeneration
import torch
from PIL import Image

model_name = "Salesforce/blip2-flan-t5-xl"

processor = Blip2Processor.from_pretrained(model_name)
model = Blip2ForConditionalGeneration.from_pretrained(model_name, torch_dtype=torch.float16).to("cuda")

img = Image.open("/content/138-1-8.webp").convert("RGB")

inputs = processor(images=img, text="Describe the traffic scene.", return_tensors="pt").to("cuda")

outputs = model.generate(**inputs)
description = processor.decode(outputs[0], skip_special_tokens=True)

print(description)
```

```python
num_cars = 5
num_buses = 2
num_trains = 1
num_pedestrians = 7
num_traffic_lights = 3

prompt = (
    f"The image shows a busy city street with {num_cars} cars, {num_buses} buses, and {num_trains} trains. "
    f"There are {num_pedestrians} pedestrians crossing the street. "
    f"The traffic lights include {num_traffic_lights} signals, with some showing red and others green. "
    "Describe the traffic scene in detail including vehicle positions, movements, pedestrian behavior, and traffic flow."
)

description = ""

for prompt in prompts:
    inputs = processor(images=img, text=prompt, return_tensors="pt").to("cuda")
    outputs = model.generate(
        **inputs,
        do_sample=True,
        temperature=0.7,
        max_new_tokens=100,
        num_beams=5
    )
    new_description = processor.decode(outputs[0], skip_special_tokens=True)
    description += new_description + "\n\n"

print(description)
```

```python
output = {
    "scene_description": scene_description,
    "model_description": description.strip(),
    "detected_objects": scene_objects,
    "license_plates": plates_info
}

import json
print(json.dumps(output, indent=2))
```
